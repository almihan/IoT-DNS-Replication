### DNS 1Second-TimeWindow Data Analysis

We analyzed DNS data from the 'dns_2019_08.pcap' file, as referenced in paper [1] . We extractted a one-day period of data. Our analysis used a time window of one second. In the graph we created, each node represents a distinct domain name. The intensity of a node's color is determined by the number of 1-second windows in which the domain appeared. Domains that appear within the same 1-second window are connected in consecutive order based on their timestamps. The intensity of an edge's color indicates the number of times that particular connection occurred across all time windows.

### Steps to create the graph

1. Group by 1-second Windows: For each row, group those that have timestamps within 1 second of each other. For each row, find the group of rows that have timestamps within 1 second of the current row.
Create a new column to store the group ID for each row.

2. Count Domain Occurrences: For each distinct domain, count how many 1-second windows it appeared in.

3. For each group, connect domains that appear in the same 1-second window, in the consecutive order based on the timestamp.
